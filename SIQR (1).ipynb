```python
import numpy as np
from scipy.optimize import fsolve
from numpy import linalg
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
T=80
N=100
temps=[T*i/N for i in range(N+1)]
```

```python
def SIQR(x,t,a,b,c,d):
    S,I,Q,R=x
    dS = -a*S*I
    dI = a*S*I-b*I-d*I
    dQ =b*I-c*Q
    dR = d*I+c*Q
    return (dS,dI,dQ,dR)
```

```python
a = 0.5
b = 0.1
c = 0.06
d = 0.06
x0 = [0.7,0.3,0.0,0.0]
Sol = odeint(SIQR,x0,temps,args=(a,b,c,d))
```

```python
sol = fsolve(SIQR,x0, args=(temps,a,b,c,d))
pt_eq = SIQR(sol,temps,a,b,c,d)
print("Point d'équilibre")
print('x=', sol)
print('f(x)=', pt_eq)
sol = sol/np.sum(sol)
eq = odeint(SIQR,sol,temps,args=(a,b,c,d))
plt.plot(temps,eq)
plt.legend(['S','I','Q,''R'])
plt.xlabel('t')
```

```python
x02 = [0.05,0.0,0.0,0.95]
sol = fsolve(SIQR,x02, args=(temps,a,b,c,d))
pt_eq = SIQR(sol,temps,a,b,c,d)
print("Point d'équilibre")
print('x=', sol)
print('f(x)=', pt_eq)
#sol = sol/np.sum(sol)
eq = odeint(SIQR,sol,temps,args=(a,b,c,d))
```

```python
plt.plot(temps,eq)
plt.legend(['S','I','Q,''R'])
plt.xlabel('t')
plt.ylim((-0.05,1.05))
```
