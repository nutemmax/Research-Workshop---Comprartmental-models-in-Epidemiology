```python
import numpy as np
from scipy.optimize import fsolve
from numpy import linalg
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
T=60
N=1000
temps=[T*i/N for i in range(N+1)]
```

```python
def SEIR(x,t,alpha, beta,gamma):
    S,E,I,R=x
    dS = -beta*S*I
    dE = beta*S*I-gamma*E
    dI = -alpha*I+gamma*E
    dR = alpha*I
    return (dS,dE,dI,dR)
```

```python
alpha = 0.5
beta = 0.2
gamma = 0.1
x0 = [0.5,0.2,0.3,0.0]
Sol = odeint(SEIR,x0,temps,args=(alpha,beta,gamma))
```

Points d'équilibres + graphe du modèle

```python
sol = fsolve(SEIR,x0, args=(temps,alpha,beta,gamma))
pt_eq = SEIR(sol,temps,alpha,beta,gamma)
print("Point d'équilibre")
print('x=', sol)
print('f(x)=', pt_eq)
sol = sol/np.sum(sol)
eq = odeint(SEIR,sol,temps,args=(alpha,beta,gamma))
plt.plot(temps,eq)
plt.legend(['S','E','I','R'])
plt.xlabel('t')
```

```python
plt.plot(temps,Sol)
plt.legend(['S','E','I','R'])
plt.xlabel('t')
plt.ylim((-0.05,1))
```
